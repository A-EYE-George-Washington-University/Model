```python
import numpy as np 
from keras.models import Model, load_model
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc, roc_auc_score
from itertools import cycle
import pandas as pd 
import matplotlib.pyplot as plt
#from sagemaker.predictor import csv_serializer 


#Call testing dataset
image_size=299
labels = []
dataset = []
x_test = np.load('/x_test.npy')
y_test = np.load('/y_test.npy')

#load model
model=load_model('/model.h5')
#model.summary()
```

```python
#test and prediction arrays
    
#change number of classes as needed
n_classes = 6

#binarize the label: one vs all
y_test1 = label_binarize(y_test, classes=[0, 1, 2, 3, 4, 5]) #change number when adding classes
y_pred = model.predict(x_test, batch_size=1, verbose=1) 
```

```python
#visualize

labels = np.array(["Normal", "Diabetes", "Cataract", "Glaucoma", "AMD", "Myopia"])

fp = dict()
tp = dict()
roc_auc = dict()

#Generate ROC curve and AUC
for i in range(n_classes):
    fp[i], tp[i], _ = roc_curve(y_test1[:, i], y_pred[:, i])
    roc_auc[i] = auc(fp[i], tp[i])

#Createt a plt figure & plot ROC curve
plt.figure(figsize=(7, 7))

colors = cycle(['aqua', 'darkorange', 'yellowgreen', 'blueviolet', 'red', 'magenta']) #change/add colors as needed
for i, color in zip(range(n_classes), colors):
    plt.plot(fp[i], tp[i], color=color,
             label='ROC curve of ' + str(labels[i]) + ' (AUC = {1:0.2f})'
             ''.format(i, roc_auc[i]))


plt.plot([0,1],[0,1], color= 'green', linestyle='--')
plt.xlim([-0.0,1.0])
plt.ylim([-0.0,1.0])
plt.xlabel('False Positive')
plt.ylabel('True Positive')
plt.legend(loc='lower right')

plt.show()
```
